# Generate Random Data using Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
```

```python
m = 0.5; c = 4
y = m*x+c
yerr = np.random.normal(0.3, 0.2, size = len(x))
```

```python
plt.errorbar(x,y, yerr)
```

```python
y1 = np.random.normal(y,0.5)
ndata = np.array([x,y1, yerr])
```

```python
plt.errorbar(x,y1, yerr, marker='o', linestyle = 'none')
```

```python
np.savetxt("data_line.txt",np.c_[x,y1,yerr], header = "x,y,sigma",delimiter= ",")
```
